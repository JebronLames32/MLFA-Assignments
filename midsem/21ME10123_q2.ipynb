```python
# importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# confused as to whether all elements should be between that range for a given X or different elements of x can be in different ranges. performed using first assumption

```python
# Choose  a  random  10D  vector  W,  where  each  value  lies  between -1  and  1.
W = np.random.uniform(-1,1,10)
# Generate  N=100 datapoints (Xi,Yi) suchthat Xi is a 10D vector whose elements lie between either (-10,-1) or (1,10)
dataset = pd.DataFrame(columns=['X','Y'])
X_list=[]
Y_list=[]
for i in range(100):
    num=np.random.randint(0,2)
    if num==0:
        X=np.random.uniform(-10,-1,10)
    else:
        X=np.random.uniform(1,10,10)
    X_list.append(X)
    # Generate  the  corresponding  output  Yi  using  the  following  equation
    Y=np.sign(np.dot(W,X))
    Y_list.append(Y)
dataset['X']=X_list
dataset['Y']=Y_list


```

```python
# Implement perceptron algorithm to find a separating hyperplane. You should not use library functions in this case.

# Initialize  the  learning  rate  to  0.01
learning_rate=0.01
# Initialize  the  number  of  iterations  to  1000
iterations=1000
# Initialize  the  number  of  misclassifications  to  1
misclassifications=1

misclassification_count=[]
# Repeat  the  following  steps  until  the  number  of  misclassifications  is  0  or  the  number  of  iterations  is  1000
while misclassifications!=0 and iterations!=0:
    misclassifications=0
    # For  each  datapoint  (Xi,Yi)  in  the  dataset
    for i in range(100):
        # If  Yi  lies on the wrong side of the hyperplane update the weight vector W
        if np.sign(np.dot(W,dataset['X'][i]))!=dataset['Y'][i]:
            W=W+learning_rate*dataset['Y'][i]*dataset['X'][i]
            misclassifications+=1
    print(misclassifications)
    misclassification_count.append(misclassifications)
    iterations-=1
print(iterations)
# Plot  the  datapoints  (Xi,Yi)  and  the  separating  hyperplane  W.T  *  X=0
x=np.linspace(-10,10,100)
y=-(W[0]*x+W[2])/W[1]
plt.plot(x,y)
for i in range(100):
    if dataset['Y'][i]==1:
        plt.scatter(dataset['X'][i][0],dataset['X'][i][1],c='r')
    else:
        plt.scatter(dataset['X'][i][0],dataset['X'][i][1],c='b')
plt.show()

```

```python
# Plot the number of updates you make in each iteration.
# The x-axis should be the iteration number and the y-axis should be the number of updates.
print(misclassification_count)
plt.plot(misclassification_count)
plt.show()
```
